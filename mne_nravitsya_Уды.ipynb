__Описание__

anime_id: Уникальный идентификатор для каждого аниме.

Name: Название аниме на языке оригинала.

English name: английское название аниме.

Other name: название на родном языке или название аниме (может быть на японском, китайском или корейском).

Score: оценка или рейтинг, присвоенный аниме.

Genres: Жанры аниме, разделенные запятыми.

Synopsis: Краткое описание сюжета аниме.

Type: Тип аниме (например, телесериал, фильм, OVA и т.д.).

Episodes: Количество серий в аниме.

Aired: Даты выхода аниме в эфир.

Premiered: сезон и год, когда состоялась премьера аниме.

Status: Статус аниме (например, Закончен показ, в настоящее время выходит в эфир и т.д.).

Producers: Продюсерские компании или продюсеры аниме.

Licensors: Лицензиары аниме (например, стриминговые платформы).

Studios: анимационные студии, которые работали над аниме.

Source: Исходный материал аниме (например, манга, лайт-новел, оригинал).

Duration: продолжительность каждого эпизода.

Rating: Возрастная категория
Ranking: Рейтинг аниме, основанный на популярности или других критериях.

Popularity: Рейтинг популярности аниме.

Favorites: Количество раз, когда аниме было отмечено пользователями как любимое.

Scored by: Количество пользователей, которые оценили аниме.

Members: Количество пользователей, которые добавили аниме в свой список на платформе.

Image URL изображения: URL-адрес изображения или постера аниме.

_Как жанр и возрастные ограничения влияют на популярность, оценки и аудиторию аниме_

```python
df.columns
```

```python
#Базарим за строки, где рейтинг анукновн ибо нах нужен
df = df.drop(df[df['Rating']=='UNKNOWN'].index, axis=0)
```

```python
print(f'Чичипирики жанров: {len(df["Genres"].unique())}')
```

```python
df = df.drop(df[df['Scored By']=='UNKNOWN'].index, axis=0)
```

```python
df = df.drop(df[df['Episodes']=='UNKNOWN'].index, axis=0)
```

```python
df = df.drop(df[df['Source']=='Unknown Source'].index, axis=0)
df = df.drop(df[df['Source']=='Unknown'].index, axis=0)
```

```python
df['Episodes'] = df['Episodes'].astype(float)
df = df.drop(df[df['Episodes']>1500].index, axis=0)
```

```python
df['Scored By'] = df['Scored By'].astype('float')
df['Score'] = df['Score'].astype('float')
```

```python
#Че че есть
df['Rating'].unique()
```

```python
plt.figure(figsize=(20,10), dpi=250)
sns.boxplot(data=df, x='Rating', y='Popularity', palette='spring', flierprops={'marker':'x'})
plt.title('Влияние ограничения на популярность')
plt.xlabel('Ограничения')
plt.ylabel('Популярность')
```

```python

plt.figure(figsize=(35,10), dpi=250)
# sns.histplot(data=df, x='Popularity', hue='Rating')
grid = sns.FacetGrid(df, col='Rating', col_wrap=2, height=10)
grid.map(sns.histplot, 'Popularity', bins=20, color='HotPink', edgecolor='white', kde=True)
grid.set_xlabels('Популярность')
#Чем больше в конце столбики, тем жанр популярнее
```

```python
df['Favorites Density'] = df['Favorites']/df['Scored By']
df['Favorites Density 2'] = df['Favorites']/df['Members']
df_grouped_favorites = df.groupby('Rating').mean(numeric_only=True)


plt.figure(figsize=(8, 5), dpi=150)
plot = sns.heatmap(data=df_grouped_favorites[['Favorites Density']], annot=True, cmap='cool')
plot.set_xlabel('Плотность добавивших в избранное (По кол-ву оценивших)')
plot.set_ylabel('Ограничения')
plt.title('Влияние ограничений, на количество оценок')
```

```python
plt.figure(figsize=(8, 5), dpi=150)
plot = sns.heatmap(data=df_grouped_favorites[['Favorites Density 2']], annot=True, cmap='spring')
plot.set_xlabel('Плотность добавивших в избранное (по кол-ву добавивших в лист просмотра)')
plot.set_ylabel('Ограничения')
plt.title('Влияние ограничений, на количество оценок')
```

```python
# sns.histplot(data=df[df['Rating']=='R+ - Mild Nudity'], x='Rating', y='Popularity')

plt.figure(figsize=(8, 5), dpi=150)
plot = sns.heatmap(data=df_grouped_favorites[['Score']], annot=True, cmap='winter')
plot.set_xlabel('Оценка')
plot.set_ylabel('Ограничения')
plt.title('Влияние ограничений на оценку')
```
